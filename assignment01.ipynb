```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

N = 10000
u = np.linspace(0,1,N)
pdf= 2*u
X = np.random.uniform(0, 1, N)
Y = np.sqrt(X)
# Plot the histogram for N = 10^4
plt.figure(figsize=(10, 6))
plt.hist(Y, bins=50, density=True, alpha=0.5, label='Histogram (N=10^4)')
plt.plot(u, pdf, label='PDF: 2u', color='blue', linewidth=2)
plt.title('Histogram vs. PDF for N=10^4')
plt.xlabel('u')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
